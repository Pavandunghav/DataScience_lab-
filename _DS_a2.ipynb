```python
##2. Scan all numeric variables for outliers. If there are outliers, use any of the suitable techniques to
#deal with them.




#Coverting all the data into numeric values


from sklearn.preprocessing import LabelEncoder



for i in data.columns:


    data[i]=LabelEncoder().fit_transform(data[i])



data
data.isna().sum()
```

```python
##Finding the outliers



def outlier_find(data_values):



  q1=np.percentile(data_values,25)
  q3=np.percentile(data_values,75)




  iqr=q3-q1



  lower_bound=q1-1.5*iqr

  upper_bound=q3+1.5*iqr



  outlier=[]

  for i in data_values:


    if i<lower_bound   or i>upper_bound:

      outlier.append(i)


  return outlier
```

```python
for i in data.columns:



  print(f"outliers in the {i} column are :{outlier_find(data[i])}")
```
