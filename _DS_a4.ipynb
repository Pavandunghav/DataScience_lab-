```python

'''
Data Analytics I
Create a Linear Regression Model using Python/R to predict home prices using Boston Housing Dataset (https://www.kaggle.com/c/boston-housing). The Boston Housing dataset contains information about various houses in Boston through different parameters. There are 506 samples and 14 feature variables in this dataset.

The objective is to predict the value of prices of the house using the given features.



'''

```

```python
##Importing necessary libraries



import numpy as np
import pandas as pd
import matplotlib.pyplot as plt




```

```python
data=pd.read_csv("/content/drive/MyDrive/DATASETS/Housing.csv")
data

```

```python
from sklearn.preprocessing import LabelEncoder

##Here we will only encode the categorical data


for i in data.columns:



 if data[i].dtypes=="object":

   data[i]=LabelEncoder().fit_transform(data[i])



data
```

```python
##Transforming the data


from sklearn.preprocessing import MinMaxScaler




data=pd.DataFrame(MinMaxScaler().fit_transform(data),columns=data.columns)


data
```

```python
##Assigning the values to x and y


y=data['price']

x=data.drop('price',axis=1)



x,y
```

```python
##Dividing the data into training and the testing



from sklearn.model_selection import train_test_split



x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)





```

```python
x_train.shape,y_train.shape


```

```python
x_test.shape,y_test.shape
```

```python
##Now model building


from sklearn.linear_model import LinearRegression


model=LinearRegression()

model.fit(x_train,y_train)
```

```python
##Making the prediction


y_pred=model.predict(x_test)

y_pred
```

```python
##Evaluation
```

```python
from sklearn.metrics import mean_absolute_error,r2_score



print(f" The Mean_absolute_error for this model is {mean_absolute_error(y_pred,y_test)}")

print(f" The r2_score_error for this model is {r2_score(y_pred,y_test)}")



```
