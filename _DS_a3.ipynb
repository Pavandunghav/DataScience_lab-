```python
##Importing teh necessary libraries


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
data=pd.read_csv("/content/drive/MyDrive/DATASETS/Iris.csv")
data
```

```python
##Statistics


def mean(feature):
  data_value=data[feature]
  addition=data_value.sum()

  count=data_value.count()



  mean=addition/count


  return mean





```

```python
def median(feature):

  data1=data[feature].sort_values()
  n=len(data[feature])

  if n//2==0:

    index=n//2



    median=(data1.iloc[index]+data1.iloc[index-1])//2



  else:
    index=n+1//2

    median=data1.iloc[index-1]



  return median








median('SepalLengthCm')
```

```python
def std(feature):



  mean_value=mean(feature)

  sum_2=0
  for i in data[feature]:


    sum_2=sum_2+(i+(mean_value)**2)



    std_value=(sum_2/2)**(-2)


    print(std_value)
    return std


print(std('SepalLengthCm'))




```
